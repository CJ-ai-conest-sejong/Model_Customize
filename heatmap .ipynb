```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt



data= pd.read_csv("/Users/soobin/Desktop/2021년 미래기술 챌린지 2 복사본/1. 이커머스 FC주문 데이터/march data.csv")
data_April = pd.read_csv("/Users/soobin/Desktop/2021년 미래기술 챌린지 2 복사본/1. 이커머스 FC주문 데이터/april data.csv")
data_May = pd.read_csv("/Users/soobin/Desktop/2021년 미래기술 챌린지 2 복사본/1. 이커머스 FC주문 데이터/may data.csv")
data_June = pd.read_csv("/Users/soobin/Desktop/2021년 미래기술 챌린지 2 복사본/1. 이커머스 FC주문 데이터/june data.csv")

data= data.dropna(axis=0, how ='any')
data_April = data_April.dropna(axis=0,how = 'any')
data_May = data_May.dropna(axis = 0, how = 'any')
data_June = data_June.dropna(axis = 0, how = 'any')
```

```python
data_org =pd.concat([data, data_April, data_May, data_June])
```

```python
data_org
```

```python
data_org['BKG_DATE']=data_org['BKG_DATE'].astype(str)
```

```python
data_org["month"]=data_org["BKG_DATE"].str[4:6]
data_org["day"]=data_org["BKG_DATE"].str[6:]
```

```python
data_org
```

```python
data_org.to_csv("join data.csv",mode='w')
```

```python
from sklearn.model_selection import train_test_split
import random

from sklearn.preprocessing import LabelEncoder

columns = ['CORP_ID','REF_ORD_NO' , 'BKG_TYP' ,'BKG_DATE', 
           'BKG_TIME' ,'SHPR_CD' ,'INV_AMT' , 'ITEM_SEQ' ,'ITEM_CD' ,
           'BRAND_NM','ITEM_QTY','ITEM_AMT',
           'ORDER_CRT_DATETIME' ,'REF_ITEM_SEQ','ORDER_IDX',
           'ORDER_YN' ,'INS_ID','INS_DATE',
           'INS_TIME','POST_ZONE', 'SPLIT_EXEC_YN','month', 'day']
dataset = data_org[columns]

dataset=dataset.dropna(axis=0, how='any')
```

```python
dataset.dtypes
```

```python
category_columns = ['CORP_ID','REF_ORD_NO','BKG_DATE','ITEM_CD','BRAND_NM','REF_ITEM_SEQ','ORDER_IDX','ORDER_YN'
                   ,'INS_ID','SPLIT_EXEC_YN','month','day']
for i in category_columns:
  L_encoder = LabelEncoder()
  L_encoder.fit(dataset[i].astype(str))
  dataset[i] = L_encoder.transform(dataset[i].astype(str))
```

```python
import seaborn as sns
import matplotlib.pyplot as plt


colormap = plt.cm.PuBu
plt.figure(figsize=(100, 98))
sns.heatmap(dataset.astype(float).corr(), linewidths = 0.1, vmax = 1.0, square = True, cmap = colormap, linecolor = "white", annot = True, annot_kws = {"size" : 48})




```
